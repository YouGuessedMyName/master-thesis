# PDR $\downarrow$

```python
from stormvogel import *
import cdd

def available_actions(s):
    if s == 0:
        return ['a','b']
    elif s == 2:
        return ['b']
    else:
        return ['a']

# TRANS = {
#     (0,'a'): [(0.5, 1), (0.5, 2)],
#     (0,'b'): [(2/3, 2), (1/3, 0)],
#     (1,'a'): [(1/2, 0), (1/2, 3)],
#     (2,'b'): [(1,0)],
#     (3,'a'): [(1,3)]
# }

P_DICT = {
    0 : {'a': {0: 0, 1: 0.5, 2: 0.5, 3: 0},
         'b': {0: 1/3, 1: 0, 2: 2/3, 3: 0}},
    1 : {'a': {0: 1/2, 1: 0, 2: 0, 3: 1/2}},
    2 : {'b': {0: 1, 1: 0, 2: 0, 3: 0}},
    3 : {'a': {0: 0, 1: 0, 2: 0, 3: 1}}}

S = [0,1,2,3]


def P(s,a,s_):
    return P_DICT[s][a][s_]

B = {3}
PROP = [1/4,1,1,1]

INIT = 0
lamb = 0.5

# def delta(s, a):
#     return TRANS[(s,a)]

def labels(s):
    return str(s)

ZETA = {0: 'a', 1: 'a', 2: 'b', 3: 'a'}
XI = {0: 'b', 1: 'a', 2: 'b', 3: 'a'}

zeta = lambda s : ZETA[s]
xi = lambda s : XI[s]

def possible_policies():
    return [zeta, xi]
```

```python
from typing import Callable
def PhiPolicy(policy: Callable, F: list[float]):
    return [
        1 if s in B else
            sum(P(s,policy(s),s_) * F[s_] for s_ in S)
        for s in S
    ]

def NextStepPolicy(policy: Callable, F: list[float]):
    return [
        sum(P(s_,policy(s_),s) * F[s_] for s_ in S)
        for s in S
    ]

def Phi(F: list[float]):
    return [
        1 if s in B else
            max([sum(P(s,a,s_) * F[s_] for s_ in S) 
                for a in available_actions(s)])
        for s in S
    ]
print(NextStepPolicy(xi, [1,0,0,0]))
```

```python
def dedup(l):
    res = []
    [res.append(x) for x in l if x not in res]
    return res

@dataclass
class LowerSet:
    eqs: list[list[float], float]

    def __contains__(self, F):
        for (row, r) in self.eqs:
            if sum(row[s] * F[s] for s in S) > r:
                return False
        return True

    def __add__(one, two):
        return LowerSet(dedup(one.eqs + two.eqs))

    def __len__(self):
        return len(self.eqs)

def downarrow(p):
    for i, entry in enumerate(p):
        if i == 0:
            assert entry != 0
        if i != 0:
            assert entry == 1
    return LowerSet([([1] + [0 for _ in range(len(p)-1)], p[0])])

def PsiPolicy(policy: Callable, G: LowerSet):
    return LowerSet([(NextStepPolicy(policy, row), r) for (row, r) in G.eqs])

def Psi(G: LowerSet):
    res = LowerSet([])
    for policy in possible_policies():
        res += PsiPolicy(policy, G)
    return res

def candidate_heuristic(p):
    return downarrow(p)

def decide_heuristic(F_k_minus_1, Gk):
    for number, policy in enumerate(possible_policies()):
        if Phi(F_k_minus_1) not in Gk:
            print("policy no.", number)
            return PsiPolicy(policy, Gk)
    return set()

def meet(l: list[list[float]]):
    if len(l) == 0:
        return []
    for li in l:
        lowest = [min(x,y) for x, y in zip(li, l[0])]
    return lowest

def conflict_heuristic_zb(F_k_minus_1, Gk):
    # First, we have to put everything in the correct format for cdd.
    # Matrix format: [b, a1, a2, a3, a4]
    # meaning: b + a1*x + a2*y + a3*z + a4*t >= 0
    assert len(Gk) == 1
    row, r = Gk.eqs[0]
    
    array = [[0] + [-x for x in row] + [-r]]
    mat = cdd.matrix_from_array(array)
    mat.rep_type = cdd.RepType.INEQUALITY
    poly = cdd.polyhedron_from_matrix(mat)
    generators = cdd.copy_generators(poly).array
    Zk = [d for d in generators if Phi(F_k_minus_1) <= d]
    meetZk = meet(Zk)
    phi_applied = Phi(F_k_minus_1)
    # print("gens", generators)
    # print("Zk", Zk)
    #print(meet(Zk))
    res = []
    for s in S:
        if row[s] != 0 and len(Zk) != 0:
            res.append(meetZk[s])
        else:
            res.append(phi_applied[s])
    return res

Fk_minus_1 = [0,0,0,0]
Gk = LowerSet(([((1,0,0,0), 1/4)]))
# print(Gk)
# print(Psi(Gk))
#print(decide_heuristic([1,1,1,1], Gk))
print(conflict_heuristic_zb(Fk_minus_1, Gk))
```

```python
MAX_ITERS = 30
def adjointPDRdown():
    F = [[], [0 for _ in S], [1 for _ in S]]
    G = []

    iteration = 0
    while True:
        Gk = G[0] if len(G) > 0 else None # index issues
        n = len(F)
        k = n - len(G)

        # DEBUG
        iteration += 1
        print(f"\n{iteration}")
        print(f"n: {n}, k: {k}")
        if iteration >= MAX_ITERS:
            print("Loopy")
            return None
        # END DEBUG

        
        for j in range(len(S)-2):
            if F[j] == F[j+1]:
                return True
        if len(G) != 0 and len(G[0]) == 0:
            return False

        # Unfold
        if len(G) == 0 and F[n-1] <= PROP:
            print("unfold")
            F.append([1 for _ in S])

        # Candidate
        elif len(G) == 0 and not F[n-1] <= PROP:
            print("candidate")
            ZZ = candidate_heuristic(PROP)
            G = [ZZ]

        # Decide
        elif len(G) > 0 and Phi(F[k-1]) not in Gk:
            print("decide")
            
            ZZ = decide_heuristic(F[k-1], Gk)
            assert F[k-1] not in ZZ
            # I don't know how to implement the other assert...
            G.insert(0, ZZ)

        # Conflict
        elif len(G) > 0 and Phi(F[k-1]) in Gk:
            print("conflict")
            z = conflict_heuristic_zb(F[k-1], Gk)
            print("z", z)
            
            assert z in Gk
            assert Phi(meet([F[k-1], z])) <= z
            
            F = [meet([Fj, z]) for (j, Fj) in enumerate(F) if j <= k] + [Fj for j in range(k+1, n-1)]
            G.pop(0)
        else:
            print("OOPS")
        print("F", F)
        print("G", G)
        
print(adjointPDRdown())
```
